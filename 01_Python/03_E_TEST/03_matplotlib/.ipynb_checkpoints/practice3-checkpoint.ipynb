```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("./tips/tips.csv")
tips = pd.DataFrame(
    data=data,
    index=data.index,
    columns=data.columns,
)
print(tips.head())
print(tips.info())
# 244行7列のデータがあることがわかる
```

```python
# 各列の要素数を数えて欠損値がないかを確認
print(tips.isnull().sum())
```

```python
# 棒グラフ plt.bar() :数の大小を比較したい
x = range(4)
x_label = tips.day.value_counts().keys()
y = tips.day.value_counts().values

plt.figure(figsize=(4,4))
plt.bar(x, y, tick_label=x_label)
plt.show()
```

```python
# 円グラフ plt.pie() :数の大小を全体に対する割合で比較したい
plt.figure(figsize=(4,4))
plt.pie(tips.day.value_counts(),
        autopct="%.1f%%",   # %.3f%%とすれば小数第3位まで表示
        labels=tips.day.value_counts().keys()
        )
plt.show()
```

```python
# 1変数の分布を確認したい
plt.figure(figsize=(8,4))
plt.hist(
    tips.total_bill, bins=20, color="green" # total_billは合計の金額。
)
plt.xlabel("total_bill")
plt.ylabel("counts")
plt.title("histogram of total bill")
plt.show()
```

```python
# 2変数の関係を確認したい
plt.figure(figsize=(8,4))
plt.scatter(x=tips["total_bill"], y=tips["tip"])
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```
